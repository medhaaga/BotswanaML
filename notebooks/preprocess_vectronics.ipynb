This notebook describes the process of creating preprocessed Vectronics data used for training models later applied on RVC data. The Vectrionics data is a raw data collected at 16 Hz frequency and we need to compute 30 second summary statistucs from this data ot coarsify it to the levels of the RVC data. We create two main data structures using the Vectronics data and behavior annotations:
  - 1. labeled Vectrnics summary data
  - 2. Unlabelled Vectronics summary data

Thje labeled Vectronics summary data is created in the following steps:
- 1.1 Match high-frequency Vectronics data with available annotations
- 1.2 Extract 30 secobd windows from the matched chunks of the vectronics data
- 1.3 Compute 9 summary statsitics based on these 30 second windows
- 1.4 Save summarized labeled Vectronics data

The unlabeled Vectronics summary data is cretaed in the following steps:
- 2.1 Load the Vectronics data for a unique pair of animal ID and date.
- 2.2 Split the data in 30 second chunks
- 2.3 Compute the 9 summary statstics for each window
- 2.4 Repeat steps 2.1-2.3 for each unqiue animal ID - date pair in the metadata
